# Analysis & Charts

```python
def claim(claim_hash):
    try:
        df_claim = pd.read_json(
            f"https://api.flipsidecrypto.com/api/v2/queries/{claim_hash}/data/latest",
            convert_dates=["BLOCK_TIMESTAMP"],
        )
    except Exception as e:
        print(e)
        return 
    return df_claim
```

```python
class DataProvider:    
    
    def __init__(self):
        self.lock = '9f7cf9ef-f7eb-45b5-92e2-dccf12d56d05'
        self.exit = '94411e01-3592-47c4-8964-c0611c1ece7a'
        self.pools = {'No LP':'No LP','-1':'No LP','1':'OSMO/ATOM',
                               '604':'STARS/OSMO',
                               '674':'DAI/OSMO',
                               '648':'PSTAKE/OSMO',
                               '3':'AKT/OSMO',
                               '704':'wETH/OSMO',
                               '605':'HUAHUA/OSMO',
                               '678':'USDC/OSMO',
                               '601':'CMDX/OSMO',
                               '560':'UST/OSMO',
                               '561':'LUNA/UST',
                               '562':'LUNA/OSMO',
                               '577':'XKI/OSMO',
                               '690':'MNTL/OSMO'}
    
    def get_lock(self):
        df = claim(self.lock)
        return df
    
    def get_exit(self):
        df = claim(self.exit)
        return df
   
```

```python
dp = DataProvider()
```

```python
df = dp.get_lock()
df
```

```python
lock_ = df.pivot(index='DAY', values='SUM(GAMMA_TOKENS)', columns='POOL_ID').fillna(0)
lock_
```

```python
df = dp.get_exit()
df
```

```python
exit_ = df.pivot(index='DAY', values='SUM(GAMMA_TOKENS)', columns='POOL_ID').fillna(0)
delta = lock_.append(-exit_)
```

```python
delta.sort_values(by='DAY').cumsum()
```
